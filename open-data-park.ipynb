```python
import pandas as pd
```

```python
df = pd.read_csv('park_data.csv',encoding = 'euc-kr')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()  #지번주소 도로명 주소 합춰줘야 할 것 같당 #주소가 있으면 위경도 정보를 얻어올 수 있지만
#시도 데이터가 없으니까 도로명/지번 주소로 시도 구군 데이터를 가져올 수 있을 것 
```

```python
df.drop(columns=['공원보유시설(운동시설)', '공원보유시설(유희시설)', '공원보유시설(편익시설)', '공원보유시설(교양시설)',
       '공원보유시설(기타시설)', '지정고시일', '관리기관명', 'Unnamed: 17'], inplace=True)
```

```python
df.head()
```

```python
df['시도'] = df['소재지도로명주소'].str.split(' ', expand=True)[0]
df['구군'] = df['소재지도로명주소'].str.split(' ', expand=True)[1]
```

```python
df.head()
```

```python
df['공원면적'] = df['공원면적'].str.replace(',', '').astype(float)
df['공원면적그래프용'] = df['공원면적'].apply(lambda x : x * 0.0001)
df['공원면적그래프용'].head()
```

```python
df.head()
```

```python
df.공원구분.value_counts().to_frame()
```

```python
from plotnine import *
```

```python
(ggplot(df)
 + aes(x='경도', y='위도')
 + geom_point()
 + theme(text=element_text(family='NanumGothic'))
)
```

```python
df = df.loc[(df['위도'] > 32 ) 
            & (df['경도'] < 132) 
            & df['시도'].notnull()]
df.shape
```

```python
(ggplot(df)
 + aes(x='경도', y='위도')
 + geom_point()
 + theme(text=element_text(family='NanumGothic'))
)
```

```python
df['소재지도로명주소'].fillna(df['소재지지번주소'],inplace = True)
```

```python
(ggplot(df)
 + aes(x='경도', y = '위도',fill = '시도')
 + geom_point()
 + theme(text=element_text(family='NanumGothic'))
)
```
